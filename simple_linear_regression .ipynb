```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
df_data = df
y_train = df_data.pop('Salary')
X_train = df_data
# df_test_predict = X_train.iloc[0:1]
lm = LinearRegression()
model = lm.fit(X_train, y_train)

# model.predict(df_test_predict)
#logistic regression
#polynomial regression
```

```python
data = df_test.iloc[0:11]
actual_salary = data.pop('Salary')

print(data)
predicted_expenses = model.predict(data)
print("Actual expenses",actual_salary)
print("predicted expenses",predicted_expenses)
```
